```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data Calculation (Price in EUR per 1 Model)
# Hunyuan: Free tier (0 EUR)
# Hyper3D: 0.5 credits/model * 0.56 EUR/credit = 0.28 EUR
# Meshy4: 20 credits/model * (3.5 EUR / 100 credits) = 0.70 EUR
# Tripo: 35 credits/model * (8.55 EUR / 1000 credits) = 0.299 EUR
# Trellis: 2 credits/model * (11.71 EUR / 3600 credits) = 0.0065 EUR

data = {
    'Tool': ['Hunyuan', 'Trellis', 'Hyper3D', 'Tripo', 'Meshy4'],
    'Price_per_Model': [0.0, 0.0065, 0.28, 0.299, 0.70]
}

df = pd.DataFrame(data).sort_values(by='Price_per_Model')

# Plotting
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']

ax = sns.barplot(x='Tool', y='Price_per_Model', data=df, palette="viridis", edgecolor='0.2')

# Add values on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.3f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points',
                fontsize=11, fontweight='bold')

plt.title('Economic Evaluation: Purchase Cost per 3D Model Generation', fontsize=15, pad=20)
plt.ylabel('Cost per Model ()', fontsize=12)
plt.xlabel('AI Generative Tool', fontsize=12)
plt.ylim(0, 0.8) # Adjusted for visibility
plt.tight_layout()
plt.savefig('economic_cost_eval.png', dpi=300)
plt.show()
```
